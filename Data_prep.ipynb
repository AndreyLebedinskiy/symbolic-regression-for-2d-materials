```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pickle
import pymatgen
import math as m
from ase.visualize.plot import plot_atoms
from pymatgen.io.ase import AseAtomsAdaptor
```

```python
#displays strcutures
def printStructure(structure):
    s0 = structure.initial_structure
    d0 = list(structure.defect_representation)
    ase_atoms = AseAtomsAdaptor.get_atoms(s0)
    fig, ax = plt.subplots()
    plot_atoms(ase_atoms, ax, radii=0.6, rotation=('10x, 0y, 0z'), show_unit_cell=True)
    xlim = 40
    ylim = 23
    ax.set_xlim(-2, xlim)
    ax.set_ylim(-2, ylim)
    ax.set_title(f'{d0}')
```

```python
structures = pd.read_pickle('data.pickle')
structures.head()
```

```python
#leave only strcutures with 2 defects 
structures['number_of_defects'] = structures.apply(lambda row: len(list(row.defect_representation)), axis=1)
twoDefect = structures.loc[structures.number_of_defects == 2]
twoDefect.head()
```

```python
#remove substitution defects
vacDefect = twoDefect.loc[twoDefect.defect_representation.apply(lambda row: row.species[0].symbol == 'X' 
                                                                and row.species[1].symbol == 'X')]
vacDefect.head()
```

```python
#extact atom types and distance between defects
def distanceCalc(defect):
    first = defect[0]
    second = defect[1]
    x = (first.frac_coords[0] - second.frac_coords[0]) ** 2
    y = (first.frac_coords[1] - second.frac_coords[1]) ** 2
    z = (first.frac_coords[2] - second.frac_coords[2]) ** 2
    distance = m.sqrt(x + y + z)
    return distance

def getDefectFirst(defect):
    x = defect[0]
    if x.frac_coords[2] > 0.3:
        return 'S2'
    elif x.frac_coords[2] < 0.2:
        return 'S1'
    else:
        return 'Mo'
    
def getDefectSecond(defect):
    x = defect[1]
    if x.frac_coords[2] > 0.3:
        return 'S2'
    elif x.frac_coords[2] < 0.2:
        return 'S1'
    else:
        return 'Mo'
    
vacDefect['first_defect'] = vacDefect.defect_representation.apply(getDefectFirst)
vacDefect['second_defect'] = vacDefect.defect_representation.apply(getDefectSecond)
vacDefect['distance'] = vacDefect.defect_representation.apply(distanceCalc)
```

```python
#Save resulting datasets
vacMo_S = vacDefect.loc[(vacDefect.first_defect == 'Mo') | (vacDefect.second_defect == 'Mo')]
vacMo_S.to_pickle("vacMo_S.pkl") 

vacSameS_S = vacDefect.loc[vacDefect.first_defect == vacDefect.second_defect]
vacSameS_S.to_pickle('vacSameS_S.pkl')

vacDifS_S = vacDefect.loc[((vacDefect.first_defect == 'S1') & (vacDefect.second_defect == 'S2')) 
              | ((vacDefect.first_defect == 'S2') & (vacDefect.second_defect == 'S1'))]
vacDifS_S.to_pickle('vacDifS_S.pkl')
```
